## 1.1  ხელით გამოთვლილი

````python
import numpy as np

# მონაცემები
x = np.array([-8.5, -8, -6, -3, -1.7, 1.2, 3.5, 5, 7, 9])
y = np.array([7, 5.5, 3.5, 4, 2, 0.9, -2.5, -3, -4, -5])

# საშუალოები
x_mean = np.mean(x)
y_mean = np.mean(y)

# ნუმერატორი
numerator = np.sum((x - x_mean) * (y - y_mean))

# დენომინატორი
denominator = np.sqrt(np.sum((x - x_mean)**2) * np.sum((y - y_mean)**2))

# კორელაციის კოეფიციენტი
r_manual = numerator / denominator

print("პირსონის კორელაციის კოეფიციენტი (ხელით გამოთვლილი):", round(r_manual, 3))
````

## 1.1 Scipy მოდულით გამოთვლილი

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# მონაცემები
x = [-8.5, -8, -6, -3, -1.7, 1.2, 3.5, 5, 7, 9]
y = [7, 5.5, 3.5, 4, 2, 0.9, -2.5, -3, -4, -5]

# კორელაციის კოეფიციენტის გამოთვლა
r, p_value = pearsonr(x, y)
print("პირსონის კორელაციის კოეფიციენტი:", round(r, 3))
````

## 1.2  განაწილების გრაფიკი

## 2.2 მონაცემთა კითხვის კოდი

````python
import pandas as pd

df = pd.read_csv("daviti_datunashvili_1_62815392.csv")
print(df.head())
````

## 2.2 sklearn-ით ლოგიგისტიკურიი რეგრესიის მოდელის შექმნა

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# მონაცემების გაყოფა
X = df.drop('is_spam', axis=1)
y = df['is_spam']

# 70% სწავლებაზე, 30% ტესტზე
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ლოგისტიკური რეგრესიის მოდელი
model = LogisticRegression()
model.fit(X_train, y_train)
````

### 2.2 მოდელის მიერ ნაპოვნი კოეფიციენტები 

````python
coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_[0]
})
print(coefficients)
````

### 2.3 გადაამოწმეთ შექმნილი მოდელი იმ მონაცემებზე, რომლებიც არ გაქვთ გამოყენებული საწვრთნელად (1 ქულა). ანგარიშში მოიყვანეთ აღრევის მატრიცა (Confusion Matrix) და აღწერეთ მისი პოვნის კოდი (1 ქულა).

### 2.3 csv ფაილიდან ტესტური მონაცემებზე შემოწმებული მოდელის სიზუზტე

````python
from sklearn.metrics import accuracy_score

# პროგნოზის გაკეთება ტესტური მონაცემების მიხედვით
y_pred = model.predict(X_test)

# სიზუსტის (accuracy) გამოთვლა
accuracy = accuracy_score(y_test, y_pred)

# შედეგების გამოტანა
print(f"მოდელის სიზუსტე: {accuracy * 100:.2f}%")
````

**მოცემული მაგალითი იქნება "სპამი", რადგან ტექსტი შეიცავს სპამისთვის დამახასიათებელ სიტყვებს, როგორიცაა "CLICK HERE", "LIMITED OFFER", "EARN MONEY", და "FAST"**

````python
from collections import Counter

def extract_features_from_text(text):
    words_list = text.split()
    word_count = Counter(word.lower().strip('!.,:;') for word in words_list)
    
    words = len(words_list)
    links = len(re.findall(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', text))
    capital_words = len([word for word in words_list if word.isupper()])
    
    spam_keywords = ['offer', 'earn', 'money', 'fast', 'limited', 'click', 'here', 'now', 'free', 'urgent', 'buy', 'discount', 'win']
    spam_word_count = sum(word_count[word] for word in spam_keywords if word in word_count)

    print(f"Words: {words}, Links: {links}, Capital Words: {capital_words}, Spam Word Count: {spam_word_count}")
    return words, links, capital_words, spam_word_count


# ფუნქცია, რომელიც ამოწმებს ელფოსტა ტექსტის მიხედვით
def check_email_spam(text, model, verbose=True):
    """
    ამოწმებს ელფოსტა ტექსტს, არის თუ არა სპამი.
    
    პარამეტრები:
    - text (str): შესამოწმებელი ელფოსტის ტექსტი
    - model: გაწვრთნილი მოდელი, რომელსაც აქვს .predict() მეთოდი
    - verbose (bool): True-ის შემთხვევაში ბეჭდავს დეტალებს

    აბრუნებს:
    - (str): "სპამი" ან "ლეგიტიმური"
    """
    # ამოღება მახასიათებლების
    features = extract_features_from_text(text)
    feature_names = ["words", "links", "capital_words", "spam_word_count"]
    
    # DataFrame შექმნა
    sample = pd.DataFrame([features], columns=feature_names)
    
    # პროგნოზირება
    prediction = model.predict(sample)[0]
    label = "სპამი" if prediction == 1 else "ლეგიტიმური"
    
    if verbose:
        print("\n📨 ტექსტი:", text)
        print("📊 მახასიათებლები:", dict(zip(feature_names, features)))
        print("🔍 კლასიფიკაცია:", label)
    
    return label


# ტექსტის შემოწმება
email_text = "CLICK HERE NOW! LIMITED OFFER! EARN MONEY FAST!!!"
result = check_email_spam(email_text, model)
print("შედეგი:", result)
````
